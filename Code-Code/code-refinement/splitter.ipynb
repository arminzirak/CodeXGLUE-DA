```python
import pandas as pd
```

```python
fixed = pd.read_csv('./data/datasets/50/fixed_labelled.csv', index_col=0)
fixed.head(1)
```

```python
buggy = pd.read_csv('./data/datasets/50/buggy_labelled.csv', index_col=0)
buggy.head(1)
```

```python
biggest_repos = buggy['repo'].value_counts()[:10]
biggest_repos
```

```python
biggest_repos.keys().values
```

```python
target_repos = biggest_repos.keys().values
```

```python
targets = fixed['repo'].map(lambda x: x in target_repos)
```

```python
splits = pd.DataFrame({'target':targets})
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
for repo in buggy['repo'].value_counts().keys():
    indices = fixed[fixed['repo'] == repo].index
    if len(indices) <= 1:
        splits.loc[indices, 'set'] = 'train'
        continue
    if repo in target_repos: 
        train, test = train_test_split(indices, test_size=0.20)   
        train, validation = train_test_split(train, test_size=0.12) 
        splits.loc[test, 'set'] = 'test'
    else:
        train, validation = train_test_split(indices, test_size=0.10)   
        
    splits.loc[train, 'set'] = 'train'
    splits.loc[validation, 'set'] = 'val'
    
    print(len(train), len(validation), len(test) if repo in target_repos else 'NA')

```
